```python
## Alpha = the size od dots get inc 
plt.scatter (x,y,c='#24EE98',alpha=.8)
```

```python
#Maker = represent the data value by any symbol
plt.scatter (x,y,c='#24EE98',alpha=.8,marker='>')
```

```python
#Grid
plt.scatter (x,y,c='#24EE98',alpha=.8,marker='>')
plt.title("this is x vs y")
plt.grid()
plt.show()
```

```python
#figure = to find the size of graph
plt.figure(figsize=(6,2))
plt.scatter(x,y,c='r',alpha=.9,marker='*')
plt.grid()
```
